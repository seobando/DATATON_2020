```python
summary = pd.read_csv("summary.csv")
summary = summary.rename(columns={"Unnamed: 0":"summary"})
summary = summary.set_index("summary")
summary.head()
```

## Get data for boxplots

```python
# Organize columns according to boxplot parameters requirement
new_summary = summary.rename(columns={'25%':'q1','50%':'med','75%':'q3','upper':'whishi','lower':'whislo','IQR':'iqr'})
new_summary = new_summary[['mean','med','q1','q3','iqr','whishi','whislo']]
new_summary["mean"] = new_summary["mean"].astype("float64")
new_summary = new_summary.round(0)
```

```python
# Setting max number of rows to display in a dataframe
pd.set_option('display.max_rows', None)
display(new_summary)
```

### Boxplot

```python
for i, var in enumerate(variables):
    result = new_summary[new_summary.index ==var]
    upper =result["whishi"].iloc[0].astype("float64")
    lower = result["whislo"].iloc[0].astype("float64")
    result = result.to_dict('records')[0]
    # Source https://stackoverflow.com/questions/62197444/how-to-create-a-boxplot-from-pre-aggregated-grouped-data-in-python
    # https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.axes.Axes.bxp.html
    fig, ax = plt.subplots(figsize=(20, 3))
    ax.bxp([result], showmeans = True, meanline = True, vert=False, showfliers = False)
    ax.set_title(var.upper())
    ax.set_yticklabels("")
```

### Get data for histograms

```python
# Initialice error array
error = []
# Run loop
for var in variables:
    try:
        # Get histogram parameters
        ## 1. Range
        summary_variable = summary[summary.index==var]
        minimum = summary_variable["min"].astype("float64").iloc[0]
        maximum = summary_variable["max"].astype("float64").iloc[0]
        step_1 = maximum - minimum
        ## 2. Number of intervals
        step_2 = 20
        ## 3. Determine the width of the intervas
        step_3 = step_1/step_2
        ## Lower range
        full_range = []
        ## Width
        width = step_3
        ## Set X
        x = 0
        ## Generate range array
        for x in range(step_2+1):
            value = minimum + width*x
            full_range.append(value)
        ## Getting classes dataframe
        classes = pd.DataFrame({"lower_range":full_range[:-1], "upper_range":full_range[1:]})
        classes = classes.round(2)
        classes["clase"] = classes["lower_range"].astype("str") + "-" + classes["upper_range"].astype("str")
        classes = classes.reset_index()
        # Set UDF to assign the class
        def clasificacion(categoria):
            categoria = int(categoria)
            for i in range(len(classes.index)):

                class_value = classes["clase"].iloc[i]
                class_range = classes["upper_range"].iloc[i]
                if categoria < class_range:
                    break

            return class_value

        clasificacion = udf(clasificacion)
        # Get test data
        path = "output/variables/" + var + ".csv"
        test = spark.read.option("header",True).option("inferschema",True).option("nullValue", "\\N").csv(path)
        # Drop null values
        test = test.na.drop(how="any")
        # Set categories according with class
        test = test.withColumn("clase",clasificacion(col("categoria")))
        # group by class
        test_classify = test.groupBy("clase").agg(sum("frecuencia_absoluta").alias("frecuencia"))
        # Convert to pandas
        test_classify_pandas = test_classify.toPandas()
        # Get index positions of original classes
        test_classify_pandas = test_classify_pandas.merge(classes, how='left', on='clase')
        # Sort classes
        test_classify_pandas = test_classify_pandas.sort_values(by="index")
        # Clean dataframe
        test_classify_pandas = test_classify_pandas.set_index("index")
        test_classify_pandas = test_classify_pandas.drop(["lower_range","upper_range"],axis=1)
        # Save
        path = "output/classifications/" + var + ".csv"
        test_classify_pandas.to_csv(path)
    except:
        # Getting missing variables
        error.append(var)
        pass     
```
